## Tipos de dados

Um tipo de dados é uma classificação que diz ao Python que tipo de informação uma variável está guardando e o que você pode fazer com ela.

Pense nos tipos de dados como recipientes específicos para guardar informações. Cada tipo de recipiente foi projetado para guardar uma coisa diferente e tem sua próprias regras de uso.

---

### A tipagem dinâmica

O mais importante em Python é que ele tem tipagem dinâmica.

Isso significa que você não precisa dizer ao Python qual é o tipo da variável. Ele descobre sozinho quando você atribiu um valor.

Pense em tipos de dados como as ferramentas certas para cada material:

- Você usa uma calculadora (operações matemáticas) para números.
- Você usa um editor de texto (juntar, recortar, colar) para textos.

O tripo de dado informa ao Python qual ferramenta usar, por exemplo, o tipo diz ao Python que faz sentido somar 5 + 10 (dois números), mas não faz sentido somar 5 + "cachorro" (um número e um texto).

-----
### Principais tipos de dados

Esses são os tipos básicos que você usará o tempo todo:

1. 'str' (string): usado para texto. Sempre estre aspas (simples ou duplas).
* EX.: "Olá, Mundo!", 'kaua.jorge@sistemafibra.org.br'.

2. 'int' (inteiro): números inteiros, sem parte decimal.
* EX.: 10, 20, 90, 2026.

3. 'float' (ponto flutuante): números que têm a parte decimal.
* EX.: 3.14, 99.9, 200.8, 10.957.
* Obs.: Ao invés do padrão brasileiro ( , ) para separar os decimais, em programação, utilizamos o ( . ).

4. 'bool' (booleano): usado para representar o verdadeiro ou falso. Essencial para tomadas de decisões (com 'if').
* EX.: True, False (note as letras maiúsculas no início).

```python
from IPython.core.prefilter import PythonOpsChecker
age = 22 # Python automaticamente entende que isso é um int
name = 'Anderson' # str (string)
balance = 10.50 # float (decimal)
alive = True # bool (booleano)

print(type(age)) # verifica o tipo

age = '22' # muda a variável age de str para int
print(type(age))

# converter str para int
age2 = int(age)
print(type(age2))

# erro
'''age = 'cachorro'
print(type(age)) # verifica o tipo
age = int(age) # tenta muydar para int
print(type(age)) # erro'''

# exemplo de int em base diferente da decimal
age3 = '33'
print(type(age3))
age4 = int(age3, 16)
print(age4)
print(type(age4))

''' As bases podem ser:
10 - decimal
16 - hexadecimal
8 - octal
2 - binária
0 - detecção automática

O Python tenta identificar a base pelo prefixo da string:

0b → binário

0o → octal

0x → hexadecimal

sem prefixo → decimal'''

int("0b1010", 0)  # 10
int("0o12", 0)    # 10
int("0xA", 0)     # 10
int("42", 0)      # 42
```

### Concatenação vs. soma

Esta é uma das confusões mais comuns (e importantes) para quem está começando. A confusão existe por motivo simples: em muitas linguagens de programação (como Python e JavaScript), o mesmo símboloe '+' é usado para fazer as duas coisas (concatenar e somar).

A diferença total está no tipo de dados que você está trabalhando.

```python
result = 1 + 1 # realiza a soma dos números inteiros
print(result)
result = "Olá " + "Mundo!" # realiza a concatenação (junção) dos textos (strings)
print(result)

name = input('Digite seu nome: ')
age = input('Digite sua idade: ')

message = 'Olá ' + name + ', você tem ' + age + ' anos de idade.'
print(message)
print(type(message)) # Obs.: a função print() sempre retorna 'str'

message2 = f'Olá {name}, você tem {age} anos de idade.' # f-string
print(message2)
```
