## Trabalhando com Variáveis

### O que é uma variável?

Uma variável é como uma caixa com um nome que você usa para guardar informaçlões (como números ou textos) na memória do computador.

```python
# Criando uma variável

idade = 22
nome = "Kauã"
humano = True

# Acessa as variáveis

print(nome, idade, humano)
```

O valor dentro dessa "caixa" pode mudar enquanto o programa está sendo executado.

```python
# Criando uma variável NOME COM VALOR "Kauã"
nome = "kauã"
print("--- Nome antes da modificação ---")
print(nome)

# Modificando valor da variável para "Fellipe"
nome = "Felippe"
print("--- Nome depois da modificação ---")
print(nome)
```

### Nomenclatura de variáveis

A nomenclatura de variáveis se divide em duas categorias: regras (o que impede o código de funcionar) e melhores práticas (o que facilita a leitura por humanos).

---

#### Regras

Estão são regras de sintaxe. Se você quebrá-la, o programa não irá funcionar ou apresentará erros. Elas variam um pouco entre as linguagens, mas estas são quase universais (e são regras do Python):

* **Não pode começar com números**: nomes como "1nome" ou "2total" são inválidos. "nome1" ou "total2" são válidos.
* **Não pode ter espaços**: 'nome do usuario' é inválido. Use 'nome_do_usuario' ou 'nomeDoUsuario'.
* **Não pode usar caracteres especiais**: a maioria dos símbolos como '!', '?', '@', '#', '%', '&' ou '-' são proibidos. O hífen é interpretado como subtração.
* **Permitido**: geralmente, você pode usar letras (a-z, A-Z), números (0-9) e o undescore ( _ ).
* **Não pode usar "palavras reservadas"**: você não pode nomear uma variável com uma palavra que a linguagem já usa para seus comandos, como 'if', 'for', 'while', 'class', 'def', etc.
* **São "case-sensitive"**: na maioria das linguagens (incluindo o Python), 'nome' é uma variável diferente de 'Nome' ou 'NOME'.

#### Boas-práticas

Essas são convenções de estilos. Quebrá-las n~çao impede o seu código de rodar, mas torna-o confuso, difícil de manter e pouco profissional.

* **O nome deve ter significado**: a variável deve descrever o que ela armazena:
|Ruim|Bom |
|:---|:---|
|'n' |'nome_usuario'|
|'x'|'contador'|
|'temp'|'temperatura'|

* **Siga o padrão de "casing" (format)**: nunca misture os padrões. Escolha um e seja consistente. O padrão geralmente depende da linguagem:
 |Padrão|Formato|Exemplo|Onde usa|
 |:---|:---|:---|:---|
 |snake_case (sublinhado)|Palavras em minúsculo separadas por underscore|'minha_variavel', 'taxa_de_juros'|É o padrão oficial (PEP 8) para variáveis e funções em Python|
 |camelCase (camelo)|Primeira palavra começa em minúscula e as seguintes começam em maiúsculas|'minhaVariavelLonga', 'taxaDeJuros'|É o padrão em JavaScript, JAVA e C#.|
 |PascalCase (Pascal)|Todas as palavras começam em maiúsculo|'MinhaVariavelLonga', 'TaxaDeJuros'|Quase universalmente usado para nomes de classes (ex: 'class ProcessadorDePagamento'|

 * **Use inglês (padrão global)**: embora você possa escrever a variável em português ('idade_do_usuario'), a prática global esmagadora é usar o inglês ('user_age').  
 Por que? Facilita a colaboração internacional, a integração com bibliotecas (que são 99% em inglês) e a busca por ajuda on-line.

 #### Pior prática

 * **Misturar os idiomas**: ex.: 'total_de_users' ou 'lista_de_clients'. Nunca faça isso.
 * **Evite nomes ambíguos**: evite letras únicas, com 'l' (parece com o número 1) ou o 'O' (parece o 0).  
 A excessão são contadores de loops simples, onde 'i', 'j', 'k' são aceitáveis por tradição (ex.: 'for i in range(10)')
 * **Evite nomes que deem informação errada**: não chame uma variável de 'lista_usuarios' se ela armazena um dicionário (mapa).

 #### Constantes
 * **Constantes em maiúsculas**: se uma variável é definida uma vez e seu valor não deve mudar (uma constante), use snake_case maiúsculo.  
 Exemplo: 'TAXA_DE_JUROS = 0.05'.
