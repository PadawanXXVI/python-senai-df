## Trabalhando com Variáveis

### O que é uma variável?

Uma variável é como uma caixa com um nome que você usa para guardar informaçlões (como números ou textos) na memória do computador.

```python
# Criando uma variável

idade = 22
nome = "Kauã"
humano = True

# Acessa as variáveis

print(nome, idade, humano)
```

O valor dentro dessa "caixa" pode mudar enquanto o programa está sendo executado.

```python
# Criando uma variável NOME COM VALOR "Kauã"
nome = "kauã"
print("--- Nome antes da modificação ---")
print(nome)

# Modificando valor da variável para "Fellipe"
nome = "Felippe"
print("--- Nome depois da modificação ---")
print(nome)
```

### Nomenclatura de variáveis

A nomenclatura de variáveis se divide em duas categorias: regras (o que impede o código de funcionar) e melhores práticas (o que facilita a leitura por humanos).
