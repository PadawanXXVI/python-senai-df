### Atividade

* Desenvolva um programa que pegue os dados (mínimo de 3) de um objeto (pessoa, produto, carro etc.) usando o `input()`
* Pegue esses dados e coloquie-os em um dicionário.
* Imprima cada valor do dicionário de forma amigável para o usuário.


```python
# Programa para cadastrar dados de uma pessoa usando input()

print(f'{" Seja bem-vindo ao sistema de cadastro de usuário! ":^80}\n'.upper())
print(f'{"** Informe os dados do usuário **":^25}\n')

# Coletando os dados
name = input("Digite o nome: ").capitalize()

# Validação da idade (somente números inteiros)
while True:
    try:
        age = int(input("Digite a idade: "))
        break
    except ValueError:
        print("Erro: digite apenas números inteiros.")

city = input("Digite a cidade de residência: ").capitalize()

gender = input(
    'Digite o gênero ("F" - para feminino ou "M" - para masculino): '
).upper()

# Convertendo a letra para o texto completo
if gender == "M":
    gender = "Masculino"
elif gender == "F":
    gender = "Feminino"
else:
    gender = "Não informado"

# Criando o dicionário
person = {
    "Name": name,
    "Age": age,
    "City": city,
    "Gender": gender
}

# Exibindo os dados de forma amigável
print(f"\n{"=== Dados cadastrados ===":^40}")
print(f"Nome: {person['Name']}")
print(f"Idade: {person['Age']} anos")
print(f"Cidade: {person['City']}")
print(f"Gênero: {person['Gender']}")

print(f"O quadrado da idade de {name} é {age**2}")
```

## Formatação de strings com `f-strings`

### O que são strings formatadas (f-strings)?

Strings formatadas, ou `f-strings`, são a maneira mais moderna, limpa e legível de inserir variáveis diretamente dentro de um texto.

Elas foram introduzidas no Python 3.6 e são hoje a forma preferida de se trabalhar com textos e variáveis.

Imagine que você tem estas variáveis:


```python
name = "Kauã"
age = 22
```

Para criar a frase "Olá, Kauã, você tem 22 anos.", você teria que fazer:


```python
message = "Olá, " + name + " , você tem " + str(age) + " anos."
print(message)
```

Problemas:

1. Feio e difícil de ler: é muito fácil se perder nos sinais '+' e nos espaços " ".
2. Erro de tipo: você não pode "somar" um texto ("olá") com um número "age". Por isso, você é obrigado a converter o número para texto usando a função `str(age)`. Isso é fácil de esquecer.


A solução é: f-string:

As `f-strings`resolvem tudo isso de forma prática.

A sintaxe é simples:

1. Coloque a letra 'f' (de "formatada") logo antes de abrir as aspas do seu texto.
2. Coloque as variáveis dentro de chaves '{}'.

#### Exemplo 1: o básico

Vamos refazer o exemplo anterior usando `f-strings`:


```python
name = "Kauã"
age = 22

message = f"Olá, {name}, você tem {age} anos."
print(message)
```

O que aconteceu?

1. O 'f' disse ao Python: "Esta string terá variáveis".
2. O Python automaticamente encontrou `{name}` e `{age}` e os substituiu pelos valores das variáveis.
3. Note que **não** precisamos usar `str(age)`, a `f-string` cuida da conversão automaticamente.


O poder extra das `f-strings`

As `f-strings` não servem apenas para concatenar diferentes tipos de dados, você pode colocar quase qualquer tipo de código Python dentro das chaves '{}'.
