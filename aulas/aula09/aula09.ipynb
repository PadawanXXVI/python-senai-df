## Entrada e saída básica: leitura de dados do usuário e impressão de resultados

### A ferramenta mais importante: a função `input()`

Em quase todos os programas que usamos, desde apps de celular, até sites, o programam precisa de informações do usuário. Pode ser um nome, uma idade, uma senha ou uma escolha de menu.

Em Python, a maneira de pusar o programa e pedir ao usuário pada digitar algo é usando a função `input()`.

### O que o `input()` faz?

A função `input()` faz três coisas muito importante:

1. Exibe uma mensagem na tela (prompt): ela mostra um texto na tela (que você escreve dentro dos parênteses) para dizer ao usuário o que ele deve digitar.

2. Espera o usuário: o programa para e fica esperando o usuário digitar alguma coisa e apertar a tecla `ENTER`.

3. Capturar o valor: qualquer coisa que o usuário digitar é capturada pelo programa para que possamos usá-la posteriormente.

### A sintaxe básica

Para usar o `input()`, você quase sempre vai querer salvar o resultado em uma variável.


```python
variavel = input("Sua mensagem para o usuário: ")
```

### Exemplo 1

Vamos desenvolver um programa que pergunta o nome do usuário e depois o cumprimenta.


```python
name = input("What's ur name? ")

print(f"Hello, {name}! Welcome to SENAI DF")
```

### O ponto mais importante sobre o `input()`

Tudo o que o `input()` retorna é um **TEXTO (string)**.

Isso é fundamental e é a maior fonte de erros para os iniciantes.

Mesmo que o usuário digite o número 18, o `input()` não entrega ao Python o inteiro 18. Ele entrega o texto "18".

### Exemplo 2: o problema dos números

Veja o que acontece se tentarmos fazer um cálculo simples com o `input()`:

```python
valorA = input("Digite o primeiro valor: ")
valorB = input("Digite o segundo valor: ")
resultado = valorA + valorB

print(f"O resultado é: {resultado}")
```
