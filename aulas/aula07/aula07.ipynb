## Matemática e operações básicas em Python

No fundo, todo computador é uma calculadora extremamente rápida. Antes de criar sistemas complexos (IAs, jogos 3D etc.) o computador faz continhas de mais e menos.

Para não ficar só na teoria, vamos aplicar isso na prática simulando o sistema financeiro de uma concessionária de veículos.

Os símbolos das 4 operações básicas são os mesmos que a gente usa no teclado:

- soma: +
- subtração: -
- multiplicação: * (asterisco em vez de 'X')
- divisão: / (barra)

```python
# Definindo os valores
#Veículos
suv_price = 120_000.00

# Adicionais
leather_seat = 5000.00

# Pagamentos
discount = 0.2
installments = 60

# Calcula o valor total do carro (soma)
total_price = suv_price + leather_seat

# Calcula o desconto do carro (multiplicação e subtração)
total_after_discount = total_price - (total_price * discount)

# Calcula o valor das parcelas (divisão)
installment_value = total_after_discount / installments

# Imprimindo a NF-e
print('--- Nota Fiscal - SENAI AUTOCAR ---')
print(f'Preço base + opcionais: {total_price: ,.2f}')
print(f'Valor final com desconto: R$ {total_after_discount: ,.2f}')
print(f'O cliente pagará {installments} parcelas de R$ {installment_value: ,.2f}')
```

## Operadores especiais

Além do básico, o Python tem três operadores que quebram um galho enorme no dia-a-da:

1. Divisão inteira (//): imagine que uma locadora tem um orçamento de 200.000 para comprar carros populares de 60.000. Quantos carros inteiros ela consegue comprar? Não dá para comprar "meio-carro". O operador // corta os decimais e devolve a parte inteira da conta.

2. Resto da divisão (%): conhecido como "módulo". Usando o exemplo acima, depois de comprar o máximo de carros possíveis, quanto dinheiro sobrouo no caixa da locadora? O % te dá exatamente o resto que sobrou da divisão.

3. Exponenciação (**): serve para calcular potência (número elevado a outro). Muito útil para calcular juros compostos em financiamentos. Em vez do acento circunflexo, no Python usamos dois asteriscos juntos.

```python
# Orçamento
budget = 200000.00
popular_price = 60000.00

# Divisão inteira (//)
purchased_cars = budget // popular_price

print(f'Com R${budget: ,.2f} é possivel comprar {int(purchased_cars)} carros.')

# Resto da divisão (%)
left_over_budget = budget % popular_price

print(f'Após a compra, vão sobrar R${left_over_budget: ,.2f} no caixa.')

# Exponenciação (**)
taxes = 1.05 # 5% ao ano
year = 4
factor = taxes ** year

print(f'Fator multiplicador dos juros após {year} anos: {factor}')
```

## Estruturas de dados

Em Python, nós temos 3 tipos de conjuntos de dados, são eles: listas, tuplas e dicionários.

Esses conjuntos são a base de como organizamos os dados.

- list (lista): pense nela como uma mochila. É uma coleção ordenada de itens, você pode adicionar, remover e trocar as coisas dentro delas.

- tuple (tupla): pense como uma caixa lacrada. É uma coleção ordenada de itens, mas depois que você fecha (cria), você nap pode mais alterar o que está dentro.

- dict (dicionário): pense nele como um catálogo de fichas (ou a lista de contatos do seu celular). Não é ordenada por posição (1,2,3), mas sim por rótulos (ou "chaves"). Você encontra um item procurando pela sua etiqueta, não pela posição.
