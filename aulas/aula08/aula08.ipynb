## 1. 'list' (listas)

Litas ~soa acoleção mais comum e flexível do Python. Elkas usam colchetes '[]' e mantêm os itens na ordem em que você os adiciona.


```python
# Criando uma lista de compras
compras = ['Detergente', 'Arroz', 'Café', 'Pneu', 'Inseticida', 'Qboa']
print(compras)
```

```python
# Acesando o
```

### Funções principais

Adicionar item (no final): .append()


```python
compras.append('WD40')

print(compras)
```

Remover um item (pelo nome, ou melhor, pelo item): .remove()


```python
compras.remove('Café')
print(compras)
```

Remover um item (pelo índice): .pop()

```python
# Removendo o primeiro item
compras.pop(0)
print(compras)

# Removendo o último item
compras.pop()
print(compras)

# Removendo de trás para frente
compras.pop(-1)
print(compras)
# Obs.: se o valor índice, no método .pop() não for informado, será removido o último item da lista
```

Modificando um item (pela posição):

```python
compras[1] = 'Laranja'
print(compras)
```

### Quando usar?

Use listas 90% do tempo. Sempre que você tiver uma coleção/conjunto de coisas e precisar adicionar, remover ou mudar itens.


## 2. 'tuple' (tuplas)

Tuplas usam parênteses '()' e são muito parecidas com listas, com um diferencial crucial: elas são imutáveis. Uma vez criadas, não podem ser alteradas.



```python
# Criando uma tupla com coordenadas (latirude, longitude)
# Coordenadas não devem mudar, por isso tupla é bom

coordenadas = (38.8977, -77.0369)
print(coordenadas)
```

```python
# Acessando o primeiro item (mesma lógica da lista)
print(coordenadas[0])
```

O que não pode fazer:
- Adcionar item: coordenadas.append(10) ➡ dá erro! (AttributeError)
- Remover item: coordenadas.remove(38.8977) ➡ dá erro! (AttributeError)
- Modificar item: coordenadas[0] = 40.0 ➡ dá erro! (TypeError)

Por quye usar algo tão limitado?

1. Segurança: garante que dados que não devem mudar (como coordenadas, datas de calendário, RGB de uma cor) permaneçam intactos.
2. Performance: são um pouco mais rápidas para o Python processar.


## 3. 'dict' (dicionários)

Dicionários são completamente diferentes. Eles não usam ordem numérica (0, 1, 2, 3 ...). Eles usam um sistema de 'chave: valor'. Você dá um 'nome' (chave) para cada item (valor). Usam chaves '{}'.

```python
# Criando um dicionário para um aluno
aluno = {
    'Nome': "Pedro",
    'Idade': 25,
    'Curso': 'Desenvolvedor Python'
}
print(f'{aluno}')

# Imprimindo por chave e valor
for k, v in aluno.items():
    print(f"{k}: {v}")
```

```python
# Acessando um item pela chave (não pela posição)
print(aluno['Nome'])
```

### Principais funções

Adicionar novo item (ou modificar um existente):

```python
# Adcionando (ou modificando um item):
aluno['Cidade'] = 'Taguatinga'
print(aluno)
```

Remover item (pela chave): del (delete)

```python
# Removendo uma chave - 'del' funciona tanto para dicionário como para listas
del aluno['Curso']

print(aluno)
```

Quando usar dicionário

Sempre que você precisar organizar dados com rótulos. Perfeito para representar objetos do mundo real (usuário, produto, carro) onde cada pedaço de informação tem um nome.

# Atividade
- Criar uma lista de materiais escolares com 10 itens
- Adicionar 2 itens nessa lista
- Atualizar 3 itens da lista
- Remover o primeiro e o penúltimo item da lista
- Obs.: A cada passo, imprimir a lista

```python
# 1) Criar uma lista de materiais escolares com 10 itens
school_materials = [
    "Caderno",
    "Lápis",
    "Borracha",
    "Caneta azul",
    "Caneta preta",
    "Régua",
    "Marca-texto",
    "Tesoura",
    "Cola",
    "Mochila"
]

print(f'{'*' * 10} Resolução da atividade {'*' * 10}')

print("\n1) Lista inicial:")
print(school_materials)


# 2) Adicionar 2 itens à lista
school_materials.append("Estojo")
school_materials.append("Papel sulfite")
# school_materials.extende(["Estojo", "Papel sulfite"]) #

print("\n2) Após adicionar 2 itens:")
print(school_materials)


# 3) Atualizar 3 itens da lista
school_materials[1] = "Lápis grafite nº2"
school_materials[5] = "Régua 30 cm"
school_materials[7] = "Tesoura sem ponta"

print("\n3) Após atualizar 3 itens:")
print(school_materials)

# 4) Remover o primeiro e o penúltimo item da lista
school_materials.pop(0)
school_materials.pop(-2)

print("\n4) Após remover o primeiro e o penúltimo item:")
print(school_materials)
```
